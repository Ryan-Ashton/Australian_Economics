# Australian Banks listed on the ASX - Quick Market Share Snap Shot

Recently I scraped a lot of publicaly available financial data for educational purposes. I wanted to start by trying something simple - finding the Market Share of the Australian Banks and see how big the "Big 4" really is...

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# The dataset I am using is a 'special' one. It required a lot of scraping and cleaning which I didn't want to include here...

# It contains the financials for all ASX listed companies

df = pd.read_csv("ASXFinancialData.csv")

#Simplify columns

df.rename(columns={"30/09/2017": "2017", "30/09/2016": "2016", "30/09/2015": "2015", "30/09/2014": "2014" }, inplace=True)

# Make sure it's a numeric datatype

df[['2017', '2016', '2015','2014']].apply(pd.to_numeric, errors='ignore')

# Show me the money (Australian Banks)

df = df.loc[(df["GICS industry group"]=="Banks") & (df["Type"]=="Total revenue")]
df.set_index('ASX Code', inplace=True, drop=True)
df.sort_values('2017', ascending=False, inplace=True)

# Let's see an ugly pie chart..

plt.figure(figsize=(50,50))
df['2017'].plot(kind='pie', sort_columns=True, fontsize=48, autopct='%1.0f%%')
plt.title("Market Share (by Revenue) for Publicaly Listed Banks in Australia - 09/16 to 09/17", size = 60)
```

## So it appears the Big 4 have ~95% of the total revenue share versus other publically listed banks.
